```python
sales_2019['Order Date'] = pd.to_datetime(sales_2019['Order Date'])

sales_data = sales_2019.groupby(['Order Date', 'Product'])['Quantity Ordered'].sum().reset_index()

# Prepare data for Prophet (one product at a time)
product_forecasts = {}
for product in sales_data['Product'].unique():
    product_data = sales_data[sales_data['Product'] == product][['Order Date', 'Quantity Ordered']]
    product_data.columns = ['ds', 'y']  # Prophet requires these column names

    from prophet import Prophet
    import matplotlib.pyplot as plt
    # Initialize Prophet model
    model = Prophet()
    model.fit(product_data)

    # Create future dates (next quarter)
    future = model.make_future_dataframe(periods=90)  # 90 days (approximately one quarter)
    forecast = model.predict(future)

    # Save forecast for the product
    product_forecasts[product] = forecast

    # Plot forecast
    model.plot(forecast)
    plt.title(f"Sales Forecast for {product}")
    plt.show()

# Analyze results
for product, forecast in product_forecasts.items():
    predicted_sales = forecast[['ds', 'yhat']].tail(90)  # Extract future predictions
    total_predicted_sales = predicted_sales['yhat'].sum()
    print(f"Predicted total sales for {product} in the next quarter: {total_predicted_sales:.2f}")
```
